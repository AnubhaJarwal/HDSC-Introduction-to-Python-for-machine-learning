```python
selected_columns = df[['Y2017', 'Area']]
grouped_data = selected_columns.groupby('Area')['Y2017'].sum()
sorted_grouped = grouped_data.sort_values().reset_index()
seventh_lowest= sorted_grouped.iloc[6]['Area']
seventh_lowest
```

```python
missing_count = df['Y2014'].isnull().sum()
total_count = len(df['Y2014'])
percentage_missing = (missing_count / total_count) * 100
print(missing_count , percentage_missing)
```

```python
group_data = df.groupby('Element')['Y2017'].sum()
group_data.get('Processing', 0)
```

```python
corr = df[['Element Code', 'Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']].corr()
least_correlated = correlations['Element Code'].idxmin()
least_correlated
```

```python
protein_supply = df[(df['Area'] == 'Madagascar') & (df['Element'] == 'Protein supply quantity (g/capita/day)')]
grouped_data = protein_supply.groupby(['Element', 'Area'])['Y2015'].sum()
protein_supply_2015 = grouped_data.get(('Protein supply quantity (g/capita/day)', 'Madagascar'), 0)
print(protein_supply_2015)
```

```python
group = df.groupby('Item')['Y2015', 'Y2018'].sum().reset_index()
wine2015 = group[group['Item'] == 'Wine']['Y2015'].values[0]
wine2018 = group[group['Item'] == 'Wine']['Y2018'].values[0]
print(wine2015 , wine2018)
```

```python
area_unique = df["Area"].unique()
len(area_unique)
```

```python
selected_columns = df[['Y2017', 'Area']]
grouped_data = selected_columns.groupby('Area')['Y2017'].sum()
highest_sum = grouped_data.idxmax()
print(highest_sum)
```

```python
year_columns = ['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']
df['Total_Variation'] = df[year_columns].sum(axis=1)
year = df[year_columns].sum().idxmax()
print(year)
```

```python
stock_variation_df = df[df['Element'] == 'Stock Variation']
grouped_data = stock_variation_df[['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']].sum()
year= grouped_data.idxmax()
print(year)
```
